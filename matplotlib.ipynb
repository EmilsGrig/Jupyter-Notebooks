```python
import matplotlib.pyplot as plt
import numpy as np
```

Imports the matplotlib pyplot extension and the numpy module. The pyplot library is used to create graphs which can be used for data analysis while numpy is used to generate numerical data.

The code below creates data values betweeen 0 and 100 (inclusive) and plots it onto a scatter graph.
- plt.scatter() can take in the parameters; alpha, c, s, marker
- alpha controls the opacity of data points and can be used to see where data overlaps
- c allows to change the colour of the plotted data
- s controls the size of each point
- marker changes the style of the plot points 

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100

plt.scatter(x_data, y_data, alpha=0.3)
```

Line graphs can be used to show the change in data over period of time using plt.plot() which takes in a list of x values and a list of y values aallowing to plot data linearly. The parameters it takes in (c, lw, linestyle) allow to change the style of the chart by changing the colour, width and the type of line.

```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83, 80, 82, 82, 83, 81, 80, 79]

plt.plot(years, weights)
```

plt.bar() allows to plot data that is divided into groups. It takes in the parameters color, align, width, edgecolor and lw.
- color allows to change the color of the bars
- align allows to cahnge the starting position of each bar
- width determines the width of each class
- edgecolor as to provide an outline to the bars while lw allows to make the outline thicker

```python
x = ["C++", "C#", "Python", "Java", "Go"]
y = [20, 50, 140, 1, 45]
plt.bar(x, y)
```

To plot data on a histogram, plt.hist() used very, where a list of normally distributed data is passed through. Parameters bins and cumulative can be passed through. The parameter bins allows to change the number of bins/bars used to represent data, it is also able to take in a list which allows set custom width for a select number of bins.

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages)
```

```python
plt.hist(ages, bins=[ages.min(), 18, 21, ages.max()])
```

```python
plt.hist(ages, cumulative=True)
```

Matplotlib can also be used to display data with a pie chart. It takes in the numerical value and a list of labels in order to create a simple pie chart. It can then use other parameters to represent fata more clearly, with methods such as explode which allows to seperate a specific value. Other parameters include autopct (allows to calculate and display a percentage of each group), pctdistance (determines the location of the percentage away from the center of the chart) and startangle (the starting positionof the data).

```python
langs = ["C++", "C#", "Python", "Java", "Go"]
votes = [20, 50, 140, 5, 45]
explodes = [0,0,0,0.1,0]
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%", startangle=90)
```

To create a box plot, a list of normally distributed data is passed through the function plt.boxplt()

```python
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
```

Box plots can be customised:

```python
first=np.linspace(0, 10, 25)
second=np.linspace(10, 200, 25)
third=np.linspace(200, 210, 25)
fourth=np.linspace(210, 230, 25)
data=np.concatenate((first, second, third, fourth))
plt.boxplot(data)
```
